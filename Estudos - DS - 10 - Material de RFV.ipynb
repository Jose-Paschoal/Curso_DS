Uma abordagem comum para usar o RFV - é usar junto de Clusterização, ou seja, iremos agrupar os clientes de acordo com características parecidas.

```python
## Alterando o formato da data - que veio como objeto no dataset

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df.info()
```

```python
## Criando uma coluna do total gasto por cliente em uma compra

df['TotalRevenue'] = df['Quantity'] * df['UnitPrice']

df.head()
```

### Primeiros Tratamentos - Retirar Valores Negativos e Nulos

```python
## excluir quantidade ou preços negativos

df = df.query('UnitPrice > 0 and Quantity >0')

## excluir CustomerID missing

df = df[df['CustomerID'].notna()]

df.shape
```

```python
## quantis e ranking

df['Quantiles'] = pd.qcut(df.TotalRevenue, 5)
df['Percentile_Rank'] = df.TotalRevenue.rank(pct=True)
df['Rank'] = df.TotalRevenue.rank()

df
```

- qcut - divide a coluna em N partes iguais, no nosso caso dividimos em 5 partes. A coluna diz em qual das partes está o cliente.


- rank - ordena a coluna, colocando 'pct=True' teremos o retorno em %, ou seja, se tivermos o valor 0.6 - quer dizer que aquele valor está acima de 60% do dataset, sem o 'pct=True' teremos a posição em número.


## Criando um Novo Dataset para Checar - Recência

```python
## selecionando duas colunas do dataset original, e agrupando elas pelo CustomerID, em seguida aplicado método 'agg' para trazer 
## a última data de compra do cliente.

dfr = (
    df[['InvoiceDate', 'CustomerID']]
    .groupby('CustomerID')['InvoiceDate']
    .agg('max')
    .reset_index())

dfr.head()
```

### Criando uma Função de Tempo para Checar a Diferença de Meses das Compras

```python
## usando a maior data como referência

c_time = dfr.InvoiceDate.max()

## função para calcular a diferença entre as datas (meses)

def month_diff(c_time, datetime_val):
    '''Calcula quantos meses entre a nossa data de referência
    e a data de última compra do usuário'''
    return 12 * (c_time.year - datetime_val.year) + (
    c_time.month - datetime_val.month)

## criano a coluna da Recência em meses

dfr['Recency'] = dfr.InvoiceDate.apply(
    lambda datetime_val: month_diff(c_time, datetime_val))

## checando

dfr.head()
```

```python
## vendo um resumo da Recency

dfr['Recency'].describe()
```

```python
## checando a distribuição

dfr['Recency'].value_counts().sort_index().plot.bar()
```

### Criando Outro Subset - Calculando F e V

```python
## separando as colunas que interessam ao cálculo

dffv = df[['CustomerID', 'InvoiceNo', 'TotalRevenue']]

## contando os pedidos por CustomerID e verificando a média de gastos nesses pedidos

dffv = (
    dffv.groupby('CustomerID')[['InvoiceNo', 'TotalRevenue']].
    agg({
        'InvoiceNo' : 'count',
        'TotalRevenue' : 'mean'
    }).reset_index())

## renomeando as colunas

dffv = dffv.rename(columns={'InvoiceNo' : 'Frequency',
                           'TotalRevenue' : 'Monetary'})

## arredondando os valores

dffv.Monetary = dffv.Monetary.round(2)

## checando o output

dffv.head()
```

```python
## unindo os dois subsets criados - e preenchendo com 0 caso necessário

df_t = dfr.merge(dffv, on='CustomerID', how='left')

df_t = df_t.fillna(0)

df_t.head()
```

## Aplicando K-Means - Algoritmos Não-Supervisionados

```python
## carregando scaling e KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

feature = ['Recency', 'Frequency', 'Monetary']

## criando cópia do dataset original

standard = df_t.copy()

## instancia o scaler

scaler = StandardScaler()

## treina o algoritmo nos dados originais

scaler.fit(df_t[feature])

## aplica a transformação e aloca as features no dataframe cópia

standard_feature = scaler.transform(df_t[feature])
standard[feature] = standard_feature

## treina com 3 clusters

kmeans = KMeans(n_clusters=3)
kmeans.fit(standard[feature])

## atribui rótulo aos indivíduos 

df_t['Cluster'] = kmeans.labels_

## verificando o dataset com o cluster adicionado

df_t.head()
```

```python
## investigando o comportamento dos clusters

_ = sns.boxplot(x='Cluster', y='Recency', data=df_t)
```

```python
## investigando o comportamento dos clusters

_ = sns.boxplot(x='Cluster', y='Frequency', data=df_t)
```

```python
## investigando o comportamento dos clusters

_ = sns.boxplot(x='Cluster', y='Monetary', data=df_t)
```

```python
## verificando as métricas por cluster

centroid = (
    df_t.groupby('Cluster')
    [['CustomerID', 'Recency', 'Frequency', 'Monetary']]
    .agg({
        'CustomerID' : 'nunique',
        'Recency' : 'mean',
        'Frequency' : 'mean',
        'Monetary' : 'mean'
    }).reset_index())

## renomeando

centroid = centroid.rename(columns={'CustomerID' : 'N_buyers'})

## ordenando

centroid.sort_values(by='Recency', ascending=True)
```

```python
## nomeando grupos - uma forma de altera

seg_map = [
    'Comprador Frequente',
    'Cliente VIP',
    'Possível Churn'
]

centroid['Segment'] = seg_map
centroid
```
