## Análise Exploratória em Python (Matplotlib)

```python
### importar a library

import matplotlib.pyplot as plt
```

```python
## gráfico padrão

plt.plot([20,30,20,40,50], 'o') ## marcador = 'o' ou 'x'

## pyplot markers - 'go-' - podendo colocar um círuclo e uma reta passado pelos dados
```

```python
plt.plot([20,30,20,40,50], 'ro-')   ## o 'r' serve para a cor red - podemos alterar o padrão
```

```python
## inserindo títulos

plt.plot([10,20,30,40],[100,200,300,400], 'ro-')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Teste de gráfico')
plt.axis([0,50,0,900]);
```

```python
## incremento

plt.plot([10,20,30,40],[100,200,300,400], 'o-', color='red')
plt.scatter([10,20,30,40] ,[500,600,600,700], color='blue')
plt.bar([10,20,30,40], [90,90,70,40], color ='green');
```

### Exemplos Práticos

```python
import pandas as pd

house = pd.read_csv('housing.csv')

house.head()
```

```python
house_median = plt.bar(x=house.groupby('ocean_proximity')['total_rooms'].mean().index,
                      height= house.groupby('ocean_proximity')['total_rooms'].mean())

plt.grid(True)
```

```python
import seaborn as sns

house_1 = pd.read_csv('housing.csv')

sns.catplot(x='ocean_proximity', kind='count', data=house_1);
```

```python
## gráfico de dispersão com regressão linear

sns.regplot(x='total_rooms', y='median_house_value', data=house_1)
```
