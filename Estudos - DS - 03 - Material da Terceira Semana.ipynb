## Estudo de Banco - Churn

### Análise para Checar Qual Perfil de Consumidor que Ocasiona Churn

```python
## importar as bilbiotecas

## bibiliotecas para manipulação de dados
import pandas as pd
import numpy as np

## bibiliotecas para análise exploratória
import matplotlib.pyplot as plt
import seaborn as sns

## configurações
plt.style.use('ggplot')
pd.set_option('display.max_rows', 20)
pd.set_option('display.max_columns',40)
```

```python
## importar dataset

df = pd.read_csv('Churn_Modelling.csv')

df.head()
```

```python
df.info()
```

```python
## checando valores únicos nas colunas
df.nunique().sort_values()
```

```python
df = df.rename(columns={'Geography' : 'Country', 'Exited' : 'Churn'})
```

```python
## checando clientes sem saldo em conta

df.query('Balance ==0')
```

```python
df.query(f'Balance in {df.Balance.nlargest(3).to_list()} or EstimatedSalary in {df.EstimatedSalary.nlargest(3).to_list()}')
```

## Análise de Dados Exploratória

```python
df.Gender.value_counts(normalize=True)
```

```python
## Distrituição de Gênero

df.Gender.value_counts(normalize=True).plot.bar()
plt.title('Distribuição de Gênero')
plt.text(-0.1, 0.55, df.Gender.value_counts(normalize=True)[0])
plt.text(0.91, 0.47, df.Gender.value_counts(normalize=True)[1]);
```

```python
## checando saldo nas contas

plt.hist(df.Balance, bins=50);
```

```python
clientes_zerados = len(df.query('Balance==0'))
total_clientes = len(df)

print(f'Temos {(clientes_zerados/total_clientes)*100}% de nossos clientes sem saldo.')
```

```python
sns.relplot(x='EstimatedSalary', y='Balance', data=df, col='Country', color='g');
```

```python
sns.jointplot(x='EstimatedSalary', y='Age', data=df);
```

## Análise de Churn

```python
_ = df.Churn.value_counts(1).plot.bar()
plt.title('Churn');
```

```python
fig, axes =plt.subplots(2, 1)

_ = sns.boxplot(data=df, x='Churn', y='EstimatedSalary', ax=axes[0])
_ = sns.boxplot(data=df, x='Churn', y='Balance', ax=axes[1])

_ = fig.tight_layout()
```

```python
df.describe().T
```

```python
df.groupby('Churn').agg({'Balance':['median', 'mean']})
```

```python
from scipy.stats import skew

agg_func = {
    'EstimatedSalary': ['min', 'median', 'mean', 'std', skew],
    'Balance': ['min', 'median', 'mean', 'std', skew],
    'Age': ['min', 'median', 'mean', 'std', skew]
}

df.groupby('Churn').agg(agg_func)
```
