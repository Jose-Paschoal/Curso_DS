## Investigando Dataset - Olist

```python
import pandas as pd
import numpy as np

from datetime import datetime
```

```python
df = pd.read_csv('olist_orders_dataset.csv')
itens = pd.read_csv('olist_order_items_dataset.csv')

df.head()
```

```python
df.shape
```

```python
itens.head()
```

```python
itens.shape
```

## Unindo dois Datasets - Merge

```python
## unindo dois datasets com chave única - exemplo de procv do excel - datasets df e itens

uniao = pd.merge(df, itens, left_on=['order_id'], right_on=['order_id'], how='left')

uniao.head()
```

```python
uniao.shape
```

### Alterando Formato de Data - de Str para Datetime

```python
## trabalhando horas - convertendo str em datatime

uniao['order_purchase_timestamp'] = pd.to_datetime(uniao['order_purchase_timestamp'])

uniao['order_estimated_delivery_date'] = pd.to_datetime(uniao['order_estimated_delivery_date'])
```

```python
uniao.info()
```

### Inserindo Nova Coluna - Calculada

```python
uniao['Prazo de Compra'] = (uniao['order_purchase_timestamp'] - uniao['order_estimated_delivery_date'])
```
