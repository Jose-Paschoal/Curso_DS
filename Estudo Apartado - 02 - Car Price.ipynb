## 02 - Separação dos Dados

```python
from sklearn.model_selection import train_test_split
```

```python
## separando os dados para tratamento via encoder no pipeline

cat_cols = []

target = 'price'

## selecionando as colunas categóricas
for col in car.columns:
    if car[col].dtype == 'object':
        cat_cols.append(col)

## selecionando colunas numéricas
num_cols = car.select_dtypes(include=[np.number]).columns.tolist()

if target in cat_cols:
    cat_cols.remove(target)
if target in num_cols:
    num_cols.remove(target)

print('Colunas categóricas:', cat_cols)
print('Colunas numéricas:', num_cols)
```

A ideia foi fazer dois tipo de separação de dados diferentes, com um loop e um select_dtypes

```python
## dividindo os dados em treino e teste

X = car.drop('price', axis=1)
y = car['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```
