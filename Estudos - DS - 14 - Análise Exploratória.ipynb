# Análise Exploratória - Dados Insurance

```python
## importando as bibliotecas

import pandas as pd
import numpy as np

## carregando o dataset

df = pd.read_csv('insurance.csv')

df
```

```python
## fuções básicas para visualização

import matplotlib.pyplot as plt

plt.scatter(data=df, x='bmi', y='charges')

plt.show()
```

```python
## histograma

plt.hist(data=df, x='age')
plt.show()
```

```python
import seaborn as sns

sns.countplot(data=df, x='smoker')
```

## Adicionando Detalhes aos Gráficos

```python
## cria uma contagem
smoke = df['smoker'].value_counts()

## tamanho do gráfico
plt.figure(figsize=(8,6))

## cria gráfico de barras com índice de contagem
plt.bar(
    smoke.index, ## valor do eixo x
    smoke.values, # valor do eixo y
    color=['steelblue', 'lightcoral'] ## cores das barras
)

## rótulo do eixo x, letra tamanho 8
plt.xlabel('Smoker', fontsize=8)

## rótulo do eixo y, letra tamanho 8
plt.ylabel('Count', fontsize=8)

## título, letra tamanho 14
plt.title('Distribuição de Fumantes', fontsize=14)

plt.show()
```
