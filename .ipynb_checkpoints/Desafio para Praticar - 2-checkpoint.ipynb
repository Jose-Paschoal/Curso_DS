```python
## visualizando a primeira questão

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## histograma no matplotlib

dados['ppage'].hist(bins=10, edgecolor = 'black')
```

```python
## histograma no seaborn

sns.histplot(dados['ppage'], kde=True, bins=10, color='blue')
```

```python
## olhando com boxplot

sns.boxplot(y=dados['ppage'], color='orange')
plt.title('Boxplot da Idade')
plt.xlabel('Idade')
plt.show()
```

## Checando o visual com Plotly Express

```python
import plotly.express as px
```

```python
fig = px.histogram(dados, x='ppage', nbins=10)

fig.show()
```
