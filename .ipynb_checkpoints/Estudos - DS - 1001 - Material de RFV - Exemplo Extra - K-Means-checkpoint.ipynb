## Criando um Subset para Estudo

```python
dfr = df[['CustomerName', 'Recency', 'Frequency', 'Monetary']]

dfr
```

```python
## criando uma classificação para os gastos

classes = [0, 50000, 75000, 100000, 250000]

labels = ['Baixo Gasto', 'Médio Gasto', 'Alto Gasto', 'Alto Padrão']

classe = pd.cut(x=dfr.Monetary, bins = classes, labels = labels)

dfr['Segment'] = classe

dfr.head()
```

```python
## quantis e ranking

dfr['Quantiles'] = pd.qcut(dfr.Monetary, 5)
dfr['Percentile_Rank'] = dfr.Monetary.rank(pct=True)
dfr['Rank'] = dfr.Monetary.rank()

dfr
```

## K-Means

```python
## carregando scaling e KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

feature = ['Recency', 'Frequency', 'Monetary']

## criando cópia do dataset original

standard = dfr.copy()

## instancia o scaler

scaler = StandardScaler()

## treina o algoritmo nos dados originais

scaler.fit(dfr[feature])

## aplica a transformação e aloca as features no dataframe cópia

standard_feature = scaler.transform(dfr[feature])
standard[feature] = standard_feature

## treina com 3 clusters

kmeans = KMeans(n_clusters=3)
kmeans.fit(standard[feature])

## atribui rótulo aos indivíduos 

dfr['Cluster'] = kmeans.labels_

## verificando o dataset com o cluster adicionado

dfr.head()
```

```python
dfr.groupby('Cluster')[['Monetary']].agg({'Monetary' : 'mean'})
```
