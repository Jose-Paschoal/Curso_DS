### 1 - Bibliotecas

```python
## bibliotecas iniciais

import pandas as pd
import numpy as np
```

```python
## alguns datasets que serão utilizados

orders = pd.read_csv('olist_orders_dataset.csv')
itens = pd.read_csv('olist_order_items_dataset.csv')
customers = pd.read_csv('olist_customers_dataset.csv')
```

### Fazendo Merge por meio da chave - customer_id

```python
## Fazendo um inner join

df = pd.merge(orders, customers, on='customer_id')

df.head()
```

```python
## Fazendo um segundo merge
```

```python
df = pd.merge(df, itens, on='order_id')

print(f'O dataset final tem {df.shape[1]} colunas e {df.shape[0]} linhas')
df.head()
```

### 2 - Fazendo Análise Descritiva do Dataset

```python
df.describe()
```

```python
df.columns
```

```python
## checagem de nulos

valores_nulos = df.isnull().sum()
valores_nulos_df = pd.DataFrame({'Variável': valores_nulos.index, 'Valores_Nulos': valores_nulos.values})

valores_nulos_df
```

```python
## outra formad de checar os nulos

import missingno
```

```python
missingno.matrix(df, figsize=(40,10))
```

```python
rows, columns = df.shape
nulos_percentual = ((df.isnull().sum() / rows) * 100).round(2)
nulos_percentual
```

```python
## checagem de valores distintos

contagem = pd.DataFrame(columns = ['Variável', 'Quantidades_distintos'])
for coluna in df.columns:
    dados = pd.DataFrame({'Variável' : [coluna], 'Quantidades_distintos' : df[coluna].value_counts().shape[0]})
    contagem = pd.concat([contagem, dados], ignore_index=True)

contagem
```
