# Análise Exploratória - Dados Insurance

```python
## importando as bibliotecas

import pandas as pd
import numpy as np

## carregando o dataset

df = pd.read_csv('insurance.csv')

df
```

```python
## fuções básicas para visualização

import matplotlib.pyplot as plt

plt.scatter(data=df, x='bmi', y='charges')

plt.show()
```

```python
## histograma

plt.hist(data=df, x='age')
plt.show()
```

```python
import seaborn as sns

sns.countplot(data=df, x='smoker')
```

## Adicionando Detalhes aos Gráficos

```python
## cria uma contagem
smoke = df['smoker'].value_counts()

## tamanho do gráfico
plt.figure(figsize=(8,6))

## cria gráfico de barras com índice de contagem
plt.bar(
    smoke.index, ## valor do eixo x
    smoke.values, # valor do eixo y
    color=['steelblue', 'lightcoral'] ## cores das barras
)

## rótulo do eixo x, letra tamanho 8
plt.xlabel('Smoker', fontsize=8)

## rótulo do eixo y, letra tamanho 8
plt.ylabel('Count', fontsize=8)

## título, letra tamanho 14
plt.title('Distribuição de Fumantes', fontsize=14)

plt.show()
```

```python
## mexendo nos contornos

## cria uma contagem
smoke = df['smoker'].value_counts()

## tamanho do gráfico
plt.figure(figsize=(8,6))

## cria gráfico de barras com índice de contagem
plt.bar(
    smoke.index, ## valor do eixo x
    smoke.values, # valor do eixo y
    color=['steelblue', 'lightcoral'] ## cores das barras
)

## rótulo do eixo x, letra tamanho 8
plt.xlabel('Smoker', fontsize=8)

## rótulo do eixo y, letra tamanho 8
plt.ylabel('Count', fontsize=8)

## título, letra tamanho 14
plt.title('Distribuição de Fumantes', fontsize=14)

## remove as linhas do topo e da direita
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

```python
smoke = df['smoker'].value_counts()

colors = ['steelblue', 'lightcoral']

fig, ax = plt.subplots(figsize=(8,6))
ax.bar(smoke.index, smoke.values, color=colors)
ax.set_xlabel('Smoke',  fontsize=8)
ax.set_ylabel('Count', fontsize=8)
ax.set_title('Distribuição de fumantes', fontsize=14)

## remove as linhas do topo e da direita
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

### Criando Gráfico de Dispersão

```python
## dispersão
fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(df['age'], df['charges'], c='red')

## cinfiguração do título e rótulo dos eixos
ax.set_title('Gráfico de Dispersão Idade vs Cobrança', fontsize=16)
ax.set_xlabel('Idade', fontsize=12)
ax.set_ylabel('Cobranças', fontsize=12)

## remoção das linhas superior e direita
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

### Criando Subplots para Comparação 

```python
## cria o local, a configuração do plot
fig, axes = plt.subplots(
    1, # número de linhas
    2, # número de colunas
    figsize=(12,6) #tamanho do plot
)

## gráfico de dispersão de BMI x Charges na posição 0
axes[0].scatter(df['bmi'], df['charges'])
axes[0].set_xlabel('BMI')
axes[0].set_ylabel('Charges')
axes[0].set_title('Charges x BMI')

## gráfico de dispersão de Age x Charges na posição 1
axes[1].scatter(df['age'], df['charges'])
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Charges')
axes[1].set_title('Charges x Age')

## ajusta o espaço entre os plots
plt.tight_layout()

## apresenta o gráfico
plt.show()
```

```python
## calcular as médias das cobranças para cada categoria - 'sex'
mean_sex = df.groupby('sex')['charges'].mean()

## calcular as médias das cobraças para cada categoria - 'region'
mean_region = df.groupby('region')['charges'].mean()

## calcular as médias das cobranças para cada categoria - 'children'
mean_children = df.groupby('children')['charges'].mean()

## calcular as médias das cobranças para cada categoria - 'smoker'
mean_smoker = df.groupby('smoker')['charges'].mean()

## cores personalizadas para as barras
colors = ['#2b8cbe', '#fe9929', '#31a354', '#a1d99b']

## criar uma figura com subplots
figs, axes =plt.subplots(2,2, figsize=(12,8))

## gráfico de barras - média de cobranças por sexo
mean_sex.plot(kind='bar', ax=axes[0,0], color=colors)
axes[0,0].set_xlabel('Sexo')
axes[0,0].set_ylabel('Cobranças Médias')
axes[0,0].set_xlabel('Cobranças Médias por Sexo')

## gráfico de barras - média de cobranças por região
mean_region.plot(kind='bar', ax=axes[0,1], color=colors)
axes[0,1].set_xlabel('Região')
axes[0,1].set_ylabel('Cobranças Médias')
axes[0,1].set_xlabel('Cobranças Médias por Região')

## gráfico de barras - média de cobranças por filhos
mean_children.plot(kind='bar', ax=axes[1,0], color=colors)
axes[1,0].set_xlabel('Filhos')
axes[1,0].set_ylabel('Cobranças Médias')
axes[1,0].set_xlabel('Cobranças Médias por Filhos')

## gráfico de barras - média de cobranças por fumante
mean_smoker.plot(kind='bar', ax=axes[1,1], color=colors)
axes[1,1].set_xlabel('Fumante')
axes[1,1].set_ylabel('Cobranças Médias')
axes[1,1].set_xlabel('Cobranças Médias por Fumantes')

## ajustar espaçamento entre os subplots
plt.tight_layout()

## exibir
plt.show()
```

```python
## fazendo gráficos pelo seaborn
import seaborn as sns
import matplotlib.pyplot as plt

## cria uma figura com 2 linhas e 2 colunas
fig, ax = plt.subplots(2,2, figsize=(12,10))

## primeira linha, primeira coluna :  gráfico de smoker
sns.countplot(data=df, x='smoker', ax=ax[0,0])
ax[0,0].set_title('Smoker')

## primeira linha, segunda coluna :  gráfico de children
sns.countplot(data=df, x='children', ax=ax[0,1])
ax[0,1].set_title('Children')

## segunda linha, primeira coluna :  gráfico de region
sns.countplot(data=df, x='region', ax=ax[1,0])
ax[1,0].set_title('Region')

## segunda linha, segunda coluna :  gráfico de sex
sns.countplot(data=df, x='sex', ax=ax[1,1])
ax[1,1].set_title('Sex')

## ajusta o layout para evitar sobreposição
plt.tight_layout()

## mostrar
plt.show()
```

### Criando uma Caixa de Texto Personalizável

```python
## cria um texto na posção 5 do eixo x e na posição 2 do eixo y, no tamanho 16
plt.text(5,2, 'Texto Exemplo', size = 16)

## definindo os limites de nossos eixos, x vai de 0 a 7 e y de 0 a 4
plt.xlim(0,7)
plt.ylim(0,4)

## mostrar
plt.show()
```

## Fazendo Gráfico com Uso da Caixa de Texto

```python
## calcular as médias de cobranças por sexo
mean_sex = df.groupby('sex')['charges'].mean()

## cores
colors = ['steelblue', 'lightcoral']

## criando figura e eixo
fig, ax = plt.subplots(figsize=(8,6))

## plota o gráfico de barras com as cores personalizadas
ax.bar(mean_sex.index, mean_sex.values, color=colors)

## configurações dos rótulos, eixos e título
ax.set_xlabel('Sexo')
ax.set_ylabel('Média das Cobranças')
ax.set_xlabel('Média das Cobranças por Sexo')

# adiciona manualmente os textos dos rótulos com tamanho de fonte maior
ax.text(0, mean_sex.values[0],
       f'{mean_sex.values[0]:.2f}', ha='center', va='bottom', fontsize=12)
ax.text(1, mean_sex.values[1],
       f'{mean_sex.values[1]:.2f}', ha='center', va='bottom', fontsize=12)

## define limites dos eixos x e y
ax.set_xlim(-0.5,1.5)
ax.set_ylim(0,17000)

## calcula a diferença percentural entre os dois sexos
diff_percent = (mean_sex['male'] - mean_sex['female']) / mean_sex['female'] * 100

## mensagem sobre a diferença salarial
ax.text(1.58, mean_sex.values[1] + 1500,
       f'Diferença Salarial de {diff_percent:.2f}%!',
       bbox={'facecolor' : 'lightsteelblue',
            'alpha':1,
            'pad':0.7,
            'edgecolor':
            'none',
            'boxstyle': 'round'},
       color = 'black',
       fontsize=11)

## mostra o gráfico
plt.show()
```
