### Clusterizaçãom - Utilizando Base Olist

Base de dados conhecida de ecommerce, com ela podemos fazer muitos estudos e desafios, e a ideia hoje é trabalhar com clusterização em cima dessa base e entender de que forma isso pode ajudar o negócio a ser mais assertivo com os clientes

link da base: https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce

```python
## importando bibliotecas

import pandas as pd
import numpy as np

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
## lendo as bases

customers = pd.read_csv('olist_customers_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
reviews = pd.read_csv('olist_order_reviews_dataset.csv')
payments = pd.read_csv('olist_order_payments_dataset.csv')
itens = pd.read_csv('olist_order_items_dataset.csv')
```

```python
## checando as bases

customers.info()
```

### Merge dos Datasets

```python
df1 = pd.merge(customers, orders, on='customer_id')

df2 = pd.merge(df1, reviews, on='order_id')

df3 = pd.merge(df2, payments, on='order_id')

df = pd.merge(df3, itens, on='order_id')

df.head()
```

```python
df.info()
```

### Criando Variáveis Novas

```python
total_compras_cliente = df.groupby('customer_unique_id')['order_id'].nunique().reset_index()

total_compras_cliente.columns = ['customer_unique_id', 'Total_Compras']

df = pd.merge(df, total_compras_cliente, on='customer_unique_id')

df.head()
```
