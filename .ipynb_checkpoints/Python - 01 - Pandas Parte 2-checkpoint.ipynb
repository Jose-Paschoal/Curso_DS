```python
def highlight_max_min(data):
    styles = pd.DataFrame('', index=data.index, columns=data.columns)
    for col in data.columns:
        max_val = data[col].max()
        min_val = data[col].min()
        styles[col] = ['background-color: lightgreen' if v == max_val else 'background-color: yellow'
                      if v == min_val else '' for v in data[col]
                      ]
        return styles

styled_df = df_pivot.style.apply(highlight_max_min, axis=None)
display(styled_df)
```

```python
## bibliotecas

import pandas as pd
import numpy as np
```

```python
## carregando dataset

bank = pd.read_csv('bank.csv', sep=';')

bank
```

```python
bank.isnull().sum()
```

```python
bank['marital'].value_counts()
```

### 1 - Olhando apenas os divorcidados, qual o percentual deles que é empreendedor? E estudantes?

```python
## primeiro vamos filtrar os divorciados

bank_filtro = bank[bank['marital']=='divorced']

bank_filtro
```

```python
bank_filtro['job'].value_counts()
```

```python
## calcular métricas

total_bank_values = (bank_filtro['job']).count()

total_bank_values_filtro = (bank_filtro['job']=='student').sum()

total_bank_values_filtro
```

```python
percentual_student = (total_bank_values_filtro / total_bank_values) * 100

print(f'O percentual de estudantes que são divorcidados é de {percentual_student:.2f}%')
```

```python
## calculando para enterpreneur

total_bank_values_filtro2 = (bank_filtro['job']=='entrepreneur').sum()

percentual_student2 = (total_bank_values_filtro2 / total_bank_values) * 100

print(f'O percentual de empreendedores que são divorcidados é de {percentual_student2:.2f}%')
```

### 2 - Criar Duas Categorias - Premium e Gold De Acordo Com Saldo - Percentil 90 e 80

```python
## calculando os percentis

percentil_80 = np.percentile(bank['balance'], 80)

percentil_90 = np.percentile(bank['balance'], 90)
```

```python
## criando uma função para atribuir uma denominação de acordo com o saldo

def atribuir_categoria(saldo):
    if saldo >= percentil_90:
        return 'premium'
    elif saldo >= percentil_80 and saldo < percentil_90:
        return 'gold'
    else:
        return 'normal'
```

```python
## aplicando a função

bank['categoria'] = bank['balance'].apply(atribuir_categoria)

bank
```

```python
bank['categoria'].value_counts()
```

### 3 - Checar de acordo com exercício 1 se existe diferenças percentuais no premium

```python
# Definir condições booleanas para marital e job
cond_marital = bank['marital'] == 'divorced'
cond_job = bank['job'] == 'entrepreneur'

# Calcular tabela de frequências cruzadas com categoria
cross_tab = pd.crosstab(cond_marital, [cond_job, bank['categoria']], normalize='index') * 100  # Normaliza por linha e multiplica por 100 para obter percentual

# Renomear índices e colunas para melhor clareza
cross_tab.index = ['Not Divorced', 'Divorced']
cross_tab.columns.set_levels(['Not Entrepreneur', 'Entrepreneur'], level=0)

# Exibir a tabela de frequências cruzadas em percentual
print("Tabela de Frequências Cruzadas (em percentual):")
print(cross_tab)
```
