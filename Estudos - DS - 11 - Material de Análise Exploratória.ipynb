```python
## importando as bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

## carrega dataset

df = pd.read_csv('housing.csv')

df.head()
```

```python
## fazendo teste com visualizações - housing_median_age

fig, axes = plt.subplots(2, 3, figsize=(12, 8))

axes[0, 0].hist(df.housing_median_age, bins=3, color='steelblue', alpha=0.7)
axes[0, 0].set_title('Histograma de 3 bins')
axes[0, 0].set_xlabel('X-axis Label')
axes[0, 0].set_ylabel('Y-axis Label')

axes[0, 1].hist(df.housing_median_age, bins=8, color='steelblue')
axes[0, 1].set_title('Histograma de 8 bins')
axes[0, 1].set_xlabel('X-axis Label')
axes[0, 1].set_ylabel('Y-axis Label')

axes[0, 2].hist(df.housing_median_age, bins=30, color='steelblue')
axes[0, 2].set_title('Histograma de 30 bins')
axes[0, 2].set_xlabel('X-axis Label')
axes[0, 2].set_ylabel('Y-axis Label')

axes[1, 0].hist(df.housing_median_age, bins=80, color='steelblue')
axes[1, 0].set_title('Histograma de 80 bins')
axes[1, 0].set_xlabel('X-axis Label')
axes[1, 0].set_ylabel('Y-axis Label')

axes[1, 1].hist(df.housing_median_age, bins=100, color='steelblue')
axes[1, 1].set_title('Histograma de 100 bins')
axes[1, 1].set_xlabel('X-axis Label')
axes[1, 1].set_ylabel('Y-axis Label')

axes[1, 2].hist(df.housing_median_age, bins=200, color='steelblue')
axes[1, 2].set_title('Histograma de 200 bins')
axes[1, 2].set_xlabel('X-axis Label')
axes[1, 2].set_ylabel('Y-axis Label')

plt.tight_layout()

plt.show()
```

```python
## fazendo teste com visualizações - median_income

fig, axes = plt.subplots(2, 3, figsize=(12, 8))

axes[0, 0].hist(df.median_income, bins=3, color='steelblue', alpha=0.7)
axes[0, 0].set_title('Histograma de 3 bins')
axes[0, 0].set_xlabel('X-axis Label')
axes[0, 0].set_ylabel('Y-axis Label')

axes[0, 1].hist(df.median_income, bins=8, color='steelblue')
axes[0, 1].set_title('Histograma de 8 bins')
axes[0, 1].set_xlabel('X-axis Label')
axes[0, 1].set_ylabel('Y-axis Label')

axes[0, 2].hist(df.median_income, bins=30, color='steelblue')
axes[0, 2].set_title('Histograma de 30 bins')
axes[0, 2].set_xlabel('X-axis Label')
axes[0, 2].set_ylabel('Y-axis Label')

axes[1, 0].hist(df.median_income, bins=80, color='steelblue')
axes[1, 0].set_title('Histograma de 80 bins')
axes[1, 0].set_xlabel('X-axis Label')
axes[1, 0].set_ylabel('Y-axis Label')

axes[1, 1].hist(df.median_income, bins=100, color='steelblue')
axes[1, 1].set_title('Histograma de 100 bins')
axes[1, 1].set_xlabel('X-axis Label')
axes[1, 1].set_ylabel('Y-axis Label')

axes[1, 2].hist(df.median_income, bins=200, color='steelblue')
axes[1, 2].set_title('Histograma de 200 bins')
axes[1, 2].set_xlabel('X-axis Label')
axes[1, 2].set_ylabel('Y-axis Label')

plt.tight_layout()

plt.show()
```

```python
df.groupby('ocean_proximity')['median_house_value'].mean()
```

```python
## criando subsets para cada opção de ocean_proximity

house_1HOCEAN = df[df['ocean_proximity']=='1H OCEAN']
house_INLAND = df[df['ocean_proximity']=='INLAND']
house_ISLAND = df[df['ocean_proximity']=='ISLAND']
house_NEARBAY = df[df['ocean_proximity']=='NEAR BAY']
house_NEAROCEAN = df[df['ocean_proximity']=='NEAR OCEAN']

## plotando os histogramas

plt.hist(house_1HOCEAN.median_income,
        bins=20,
        color='steelblue',
        alpha=0.7,
        label='1H OCEAN')

plt.hist(house_INLAND.median_income,
        bins=20,
        color='orange',
        alpha=0.7,
        label='INLAND')

plt.hist(house_ISLAND.median_income,
        bins=20,
        color='red',
        alpha=0.7,
        label='ISLAND')

plt.hist(house_NEARBAY.median_income,
        bins=20,
        color='green',
        alpha=0.7,
        label='NEAR BAY')

plt.hist(house_NEAROCEAN.median_income,
        bins=20,
        color='purple',
        alpha=0.7,
        label='NEAR OCEAN')

## rótilos dos eixos e título do gráfico

plt.xlabel('median_income')
plt.ylabel('Frequency')
plt.title('Histograma de Income por Ocean Proximity')

## incluir legenda
plt.legend()

## exibe o gráfico
plt.show()
```

## CDP - Cumulative Distribution Plot - Alternativa ao Histograma

```python
## distribuição acumulada

x = np.sort(df.housing_median_age)
y = np.arange(len(df))/float(len(df))

plt.xlabel('Housing Median Age')
plt.ylabel('% Acumulado')

plt.title('Gráfico de Distribuição Acumulada para Median Age')

plt.plot(x, y, marker='o')
```
